```python
# Wikipedia API Gathering

# Import necessary libraries
import wikipedia
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
# Read Wikipedia pages
# Excluding Databricks because Wikipedia cannot ID it

# Read 6 companies' pages
home_depot=wikipedia.page(title="Home Depot")
bms=wikipedia.page(title='Bristol Myers Squibb')
apple=wikipedia.page(title='Apple Inc')
goldman_sachs=wikipedia.page(title='Goldman Sachs')
accenture=wikipedia.page(title='Accenture plc')

# Print one of the pages
print(home_depot.content)
```

```python
# Visualize results in several word clouds

# Create stopwords list
stopwords_ls = ["Goldman", "Sachs", "Home", "Depot",
             "Bristol", "Myers", "Squibb",
             "Accenture", "Apple"] + list(STOPWORDS)

# Create generate_word_cloud function
# Modified from Prof. Hickman's code and https://towardsdatascience.com/simple-wordcloud-in-python-2ae54a9f58e5


# Home Depot
def generate_word_cloud(text):

    # Define a function to plot word cloud
    def plot_cloud(wordcloud):
        # Set figure size
        plt.figure(figsize=(40, 30))
        # Display image
        plt.imshow(wordcloud) 
        # No axis details
        plt.axis("off");

    # Generate word cloud
    wordcloud = WordCloud(
        max_font_size = 800,
        max_words = 250,
        background_color = "orange",
        width = 3000,
        height = 2000, 
        random_state = 1, 
        colormap = "Pastel2", 
        collocations = False,
        stopwords = stopwords_ls
        ).generate(text)
    plot_cloud(wordcloud)
    
    # Save word cloud
    wordcloud.to_file("../501-project-website/images/Home_Depot_Wordcloud.jpg")
    
generate_word_cloud(home_depot.content)
```
