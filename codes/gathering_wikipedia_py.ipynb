```python
# Wikipedia API Gathering

# Import necessary libraries
import wikipedia
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
# Read Wikipedia pages

# Select which pages to read
page=wikipedia.page(title=["Home Depot", "Databricks", "Bristol Myers Squibb", "Accenture", "Apple", "Goldman Sachs"])

# Print content
print(page.content)
```

```python
# Visualize results in word cloud

# Create stopwords list
stopwords_ls = ["Goldman", "Sachs", "Home", "Depot",
             "Databricks", "Bristol", "Myers", "Squibb",
             "Accenture", "Apple"] + list(STOPWORDS)

# Create generate_word_cloud function
# Modified from Prof. Hickman's code and https://towardsdatascience.com/simple-wordcloud-in-python-2ae54a9f58e5
def generate_word_cloud(text):

    # Define a function to plot word cloud
    def plot_cloud(wordcloud):
        # Set figure size
        plt.figure(figsize=(40, 30))
        # Display image
        plt.imshow(wordcloud) 
        # No axis details
        plt.axis("off");

    # Generate word cloud
    wordcloud = WordCloud(
        max_font_size = 800,
        max_words = 250,
        background_color = "gray",
        width = 3000,
        height = 2000, 
        random_state = 1, 
        colormap = "Pastel2", 
        collocations = False,
        stopwords = stopwords_ls
        ).generate(text)
    plot_cloud(wordcloud)
    
    # Save word cloud
    wordcloud.to_file("../501-project-website/images/Wiki_Wordcloud.jpg")
    
generate_word_cloud(page.content)
```
